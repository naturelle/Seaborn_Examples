```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
iris = sns.load_dataset("iris")
iris.dtypes
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips,hue="day", size="size", palette="YlGnBu")
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x="sex", y="tip",data=tips, palette="YlGnBu",hue="day" )
```

```python
sns.boxplot(x="day", y="total_bill",data=tips,palette="YlGnBu",hue="sex")
```

```python
sns.stripplot(x="day", y="tip",data=tips,palette="YlGnBu",hue="sex", dodge=True)
```

```python
sns.jointplot(x="tip", y="total_bill",data=tips,palette="YlGnBu", kind="reg")
```

```python
sns.jointplot(x="tip", y="total_bill",data=tips,palette="YlGnBu", kind="kde", fill=True, cmap="YlGnBu")
```

```python
sns.jointplot(x="tip", y="total_bill",data=tips,palette="YlGnBu", kind="hex", cmap="YlGnBu")
```

```python
sns.jointplot(x="tip", y="total_bill",data=tips,palette="YlGnBu",hue="sex")
```

```python
sns.pairplot(titanic.select_dtypes(['number']),hue="pclass")
```
